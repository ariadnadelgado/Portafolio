```python
"Ejercicio 4: Given an array of integers, find the pair of adjacent elements that has the largest product and return that product."
def solution(inputArray):
    high_product = float('-inf')
    for i in range(len(inputArray) - 1):
        product = inputArray[i] * inputArray[i + 1]
        if product > high_product:
            high_product = product
    return high_product
```

```python
"Ejercicio 7: Given a sequence of integers as an array, determine whether it is possible to obtain a strictly increasing sequence by removing no more than one element from the array."
def solution(sequence):
    droppped = False
    last = prev = min(sequence) - 1
    for elm in sequence:
        if elm <= last:
            if droppped:
                return False
            else:
                droppped = True
            if elm <= prev:
                prev = last
            elif elm >= prev:
                prev = last = elm
        else:
            prev, last = last, elm
    return True
```

```python
"Ejercicio 9: Given an array of strings, return another array containing all of its longest strings."
def solution(inputArray):
    max_length = 0
    longest_string = []
    for string in inputArray:
        length = len(string)
        if length > max_length:
            max_length = length
            longest_string = [string]
        elif length == max_length:
            longest_string.append(string)
    return longest_string
```

```python
"Ejercicio 10: Given two strings, find the number of common characters between them."
def solution(s1, s2):
    common_char = 0
    for char in s1:
        if char in s2:
            common_char +=1
            s2 = s2.replace(char, '', 1) #remueve el char que se repite para evitar contar duplicados
    return common_char
```

```python
"Ejercicio 11: Ticket numbers usually consist of an even number of digits. A ticket number is considered lucky if the sum of the first half of the digits is equal to the sum of the second half."
"Given a ticket number n, determine if it's lucky or not."
def solution(n):
    n_string = str(n)
    longitud = len(n_string)

    #verifica si la longitud del numero es par
    if longitud % 2 != 0:
        return False

    #calcular suma de primera mitad
    pm_suma = sum(int(digit) for digit in n_string[:longitud // 2])
    #calcular suma de la segunda mitad
    sm_suma = sum(int(digit) for digit in n_string[longitud // 2:])

    return pm_suma == sm_suma
```

```python
"Ejercicio 12: Some people are standing in a row in a park. There are trees between them which cannot be moved. Your task is to rearrange the people by their heights in a non-descending order without moving the trees. People can be very tall!"
def solution(a):
    #Saca las alturas del array
    heights = [height for height in a if height != -1]

    #ordena la lista
    heights.sort()

    #reeemplaza los elemsntos que no son arboles con alturas ordenadas
    j = 0
    for i in range(len(a)):
        if a[i] != -1:
            a[i] = heights[j]
            j += 1
    return a
```

```python
"Ejercicio 13: Write a function that reverses characters in (possibly nested) parentheses in the input string."
"Input strings will always be well-formed with matching ()s."

def solution(inputString):
    stack = []
    result = ""

    for char in inputString:
        if char == '(':
            stack.append('')
        elif char == ')':
            substring = stack.pop()[::-1]
            if stack:
                stack[-1] += substring
            else:
                result += substring
        else:
            if stack:
                stack[-1] += char
            else:
                result += char
    return result
```

```python
"Ejercicio 14: Several people are standing in a row and need to be divided into two teams. The first person goes into team 1, the second goes into team 2, the third goes into team 1 again, the fourth into team 2, and so on."
"You are given an array of positive integers - the weights of the people. Return an array of two integers, where the first element is the total weight of team 1, and the second element is the total weight of team 2 after the division is complete."

def solution(a):
    teamone = 0
    teamtwo = 0

    for i, weight in enumerate(a):
        if i % 2 == 0:
            teamone += weight
        else:
            teamtwo += weight

    return [teamone, teamtwo]
```

```python
"Ejercicio 15: Given a rectangular matrix of characters, add a border of asterisks(*) to it."

def solution(picture):
    border = ['*' + row + '*' for row in picture] #que tipo de simbolo se implementara para el borde

    border_length = len(border[0]) #longitud del borde

    top_border = '*' * border_length #llenado de los bordes superiores

    border.insert(0, top_border)
    border.append(top_border)

    return border
```
